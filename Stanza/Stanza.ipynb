```python
import stanza
import pandas as pd
import spacy
from spacy import displacy
```

```python
stanza.download(lang ='fr', model_dir='.../stanza_models')
nlp_fr = stanza.Pipeline(lang ='fr', dir ='.../stanza_models')
```

```python
stanza.download(lang ='en', model_dir='.../stanza_models')
```

```python
nlp_en = stanza.Pipeline(lang ='en', dir ='.../stanza_models')
```

```python
df = pd.read_excel('7000 sentences Corpus with IDs.xlsx', index_col=0)
```

```python
df
```

```python
en = pd.DataFrame(df, columns = ['English'])
```

```python
nouns=[]
for sen in en['English'][:100]:
    dictt= nlp_en(sen).to_dict()[0]
    noun=''
    for i in dictt:
        if 'upos' in i:
            if i['upos']=='NOUN':
                if i['text'] in sen:
                    noun=(i['text'])
    nouns.append(noun)              
```

```python
len(nouns)
```

```python
data_en = en[0:100]
data_en['Noun']=nouns
data_en[:5]
```

```python
verbs=[]
for sen in en['English'][:100]:
    dictt= nlp_en(sen).to_dict()[0]
    verb=''
    for i in dictt:
        if 'upos' in i:
            if (i['upos']=='VERB' or i['upos']=='AUX'):
                if i['text'] in sen:
                    verb=(i['text'])
    verbs.append(verb)
```

```python
verbs[:5]
```

```python
data_en['Verb']=verbs
data_en[:5]
```

```python
Question = []
for i in data_en.English[1].split():
    if i == data_en.Verb[1]:
        Question.append('...')
    else:
        Question.append(i)
```
